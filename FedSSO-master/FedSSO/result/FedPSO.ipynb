## CIFAR10

### FedAVG

```python
df_AVG = []
```

```python
df_AVG_0 = pd.read_csv('FedAvg_CIFAR10_randomDrop_0%_output_C_0.5_LR_0.0025_CLI_10_CLI_EPOCHS_5_TOTAL_EPOCHS_30_BATCH_10.csv', header=None)
df_AVG_0 = df_AVG_0.rename(columns={0: "loss", 1: "accuracy"})
df_AVG.append(df_AVG_0)
```

```python
df_AVG_1 = pd.read_csv('FedAvg_cifar10_randomDrop_0%_output_C_0.5_LR_0.0025_CLI_10_CLI_EPOCHS_5_TOTAL_ROUND_30_BATCH_10_1.csv', header=None)
df_AVG_1 = df_AVG_1.rename(columns={0: "loss", 1: "accuracy"})
df_AVG.append(df_AVG_1)
```
